```python
import random
import time
def math_game():
    penalty = 3
    while True:
        num1 = random.randint(1, 100)
        num2 = random.randint(1, 100)
        correct_sum = num1 + num2
        user_answer = int(input(f"Новый пример: {num1} + {num2}?\n"))
        if user_answer == correct_sum:
            print("Поздравляю! Ваш ответ правильный")
            break
        else:
            print(f"Неверно! Штрафная пауза: {penalty} секунды")
            time.sleep(penalty)
            penalty += 1
```
